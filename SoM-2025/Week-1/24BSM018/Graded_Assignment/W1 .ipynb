### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
ghost_pokemon = week1File[(week1File["Type 1"] == "Ghost") & (week1File["Legendary"] == False) & (week1File["Generation"] == 4)]
pokemon_sort = ghost_pokemon.sort_values(by="Speed", ascending=False)
second_fastest_ghost =pokemon_sort.iloc[1]
print(second_fastest_ghost)
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
charizard_stats = week1File.loc[week1File["Name"] == "Charizard", ["Attack", "Defense"]]
charizard_attack = charizard_stats["Attack"].values[0]
charizard_defense = charizard_stats["Defense"].values[0]
strong_defense_weak_attack = week1File[
    (week1File["Legendary"] == False) &
    (week1File["Defense"] > charizard_defense) &
    (week1File["Attack"] < charizard_attack)
]

count_pokemon = len(strong_defense_weak_attack)

print(f"Number of non-legendary Pokmon with stronger Defense but weaker Attack than Charizard: {count_pokemon}")
```
